```python
import pandas as pd
import bar_chart_race as bcr
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Video
```

```python
data = pd.read_csv('./data/pib_per_capita_countries_dataset.csv')
data
```

```python
data = data[['country_name', 'year', 'gdp_per_capita']]
```

## Filtering data only aouth america

```python
paises_suramerica = [
    "ARGENTINA", "BOLIVIA", "BRASIL", "CHILE", "COLOMBIA", 
    "ECUADOR", "GUYANA", "PARAGUAY", "PERU", "SURINAM", 
    "URUGUAY", "VENEZUELA"
]

df_data_sur_america = data[data['country_name'].isin(paises_suramerica)]
```

### Wide Table Format Data

```python
data = data.pivot(index='year', columns='country_name', values='gdp_per_capita')
df_data_sur_america = df_data_sur_america.pivot(index='year', columns='country_name', values='gdp_per_capita')
```

```python
data
```

### Bar Chart race of GDP Per Capita fo the all countries of the world

```python
bcr.bar_chart_race(df=data,
                   filename='gdp_per_capita_evolution.mp4',
                   title='GDP Per Capita Evolution from 1960',
                   orientation='h',
                   sort='desc',
                   n_bars=10,
                   fixed_order=False,
                   fixed_max=False,
                   steps_per_period=10,
                   interpolate_period=False,
                   label_bars=True,
                   bar_size=0.9,
                   period_label={'x': .99, 'y': .25, 'ha': 'right', 'va':'center'},
                   period_fmt='{x:.0f}',
                   period_summary_func=lambda v, r: {'x': .99, 'y': .18, 's': '', 'ha': 'right', 'size': 20, 'family': 'Courier New'},
                   period_length=1000,
                   figsize=(8,6),
                   dpi=500,
                   cmap='Plotly',
                   title_size=16,
                   bar_label_size=12,
                   tick_label_size=10,
                   shared_fontdict={'color': '.1'},
                   scale='linear',
                   writer=None,
                   fig=None,
                   bar_kwargs={'alpha': .7},
                   filter_column_colors=True)
```

```python
bcr.bar_chart_race(df=df_data_sur_america,
                   filename='gdp_per_capita_sur_america_evolution.mp4',
                   title='South America GDP Per Capita evolution from 1960 ranking',
                   orientation='h',
                   sort='desc',
                   n_bars=10,
                   fixed_order=False,
                   fixed_max=False,
                   steps_per_period=10,
                   interpolate_period=False,
                   label_bars=True,
                   bar_size=0.9,
                   period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
                   period_fmt='{x:.0f}',
                   period_summary_func=lambda v, r: {'x': .99, 'y': .18,'s': '',
                     'ha': 'right', 'size': 20, 'family': 'Courier New'},
                   period_length=1200,
                   figsize=(8,6),
                   dpi=250,
                   cmap='Plotly',
                   title_size=16,
                   bar_label_size=12,
                   tick_label_size=10,
                   shared_fontdict={'color' : '.1'},
                   scale='linear',
                   writer=None,
                   fig=None,
                   bar_kwargs={'alpha': .7},
                   filter_column_colors=True)
```
