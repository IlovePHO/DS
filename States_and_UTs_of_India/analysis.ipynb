```python
plt.figure(figsize=(10,5))
sns.barplot(data=data, x='Name', y='Population', palette="husl")
plt.title('State & UT Population of India')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=data, x='Name', y='Area (sq. km)', palette="husl")
plt.title('State & UT Area (sq. km) of Inida')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.show()
```

```python
new_states=data[(data['year']>=2000) & (data['Type'].str.contains('State'))][['Name']]
print(f'New States :{len(new_states)} ')
print(new_states)
```

```python
new_ut=data[(data['year']>=2000) & (data['Type'].str.contains('Union Territories'))][['Name']]
print(f'New UT :{len(new_ut)} ')
print(new_ut)
```

```python
# df_state=df[df['Type'].str.contains('State')]
data_state=data[data['Type']=='State']
data_state.head()
```

```python
# df_ut=df[df['Type'].str.contains('Union Territories')]
data_ut=data[data['Type']!='State']
data_ut.head()
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
sns.barplot(data=data_state, x='Name', y='Population', palette="husl")
plt.title('State wise Population of Inida')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
sns.barplot(data=data_ut, x='Name', y='Population', palette="husl")
plt.title('Union Territories wise Population of Inida')
plt.xlabel('Union Territories')
plt.ylabel('Population')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
sns.barplot(data=data_state, x='Name', y='Area (sq. km)', palette="husl")
plt.title('State wise Area (sq. km) of Inida')
plt.xlabel('State')
plt.ylabel('Area (sq. km)')
plt.xticks(rotation=90)

plt.subplot(1, 2, 2)
sns.barplot(data=data_ut, x='Name', y='Area (sq. km)', palette="husl")
plt.title('Union Territories wise Area (sq. km) of Inida')
plt.xlabel('Union Territories')
plt.ylabel('Area (sq. km)')
plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(20, 15))

plt.subplot(2, 2, 1)
sns.barplot(data=data_state, x='Name', y='Population', palette="husl")
plt.title('State wise Population of Inida')
plt.xlabel('State')
plt.ylabel('Population')
plt.xticks(rotation=90)

plt.subplot(2, 2, 2)
sns.barplot(data=data_ut, x='Name', y='Population', palette="husl")
plt.title('Union Territories wise Population of Inida')
plt.xlabel('Union Territories')
plt.ylabel('Population')
plt.xticks(rotation=90)

plt.subplot(2, 2, 3)
sns.barplot(data=data_state, x='Name', y='Area (sq. km)', palette="husl")
plt.title('State wise Area (sq. km) of Inida')
plt.xlabel('State')
plt.ylabel('Area (sq. km)')
plt.xticks(rotation=90)

plt.subplot(2, 2, 4)
sns.barplot(data=data_ut, x='Name', y='Area (sq. km)', palette="husl")
plt.title('Union Territories wise Area (sq. km) of Inida')
plt.xlabel('Union Territories')
plt.ylabel('Area (sq. km)')
plt.xticks(rotation=90)

plt.show()
```

```python
data_zone_count=data.groupby('Zone')['Name'].count()
print(data_zone_count)
data_zone_count.plot(kind='bar')
```

```python
data_zone_central=data[data['Zone']=='Central']
data_zone_central
```

```python
data_zone=data['Zone'].unique().tolist()
data_zone
```

```python
plt.figure(figsize=(15, 20))

for i, zone in enumerate(data_zone):
    plt.subplot(3, 2, i+1)
    sns.barplot(data=data[data['Zone']==zone], x='Name', y='Population', palette="husl")
    plt.title(f' {zone} Zone Population')
    plt.xlabel(f'{zone} States')
    plt.ylabel('Population')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 20))

for i, zone in enumerate(data_zone):
    plt.subplot(3, 2, i+1)
    sns.barplot(data=data[data['Zone']==zone], x='Name', y='Area (sq. km)', palette="husl")
    plt.title(f' {zone} Zone Area (sq. km)')
    plt.xlabel(f'{zone} States')
    plt.ylabel('Area (sq. km)')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
